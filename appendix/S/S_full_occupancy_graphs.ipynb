```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
with open('data/500K/Sn/sites.dat', 'r') as inf:
    sn500K = eval(inf.read())
    
with open('data/600K/Sn/sites.dat', 'r') as inf:
    sn600K = eval(inf.read())
    
with open('data/700K/Sn/sites.dat', 'r') as inf:
    sn700K = eval(inf.read())
    
with open('data/800K/Sn/sites.dat', 'r') as inf:
    sn800K = eval(inf.read())
    
with open('data/900K/Sn/sites.dat', 'r') as inf:
    sn900K = eval(inf.read())
    
with open('data/1000K/Sn/sites.dat', 'r') as inf:
    sn1000K = eval(inf.read())


with open('data/500K/Ge/sites.dat', 'r') as inf:
    ge500K = eval(inf.read())
    
with open('data/600K/Ge/sites.dat', 'r') as inf:
    ge600K = eval(inf.read())
    
with open('data/700K/Ge/sites.dat', 'r') as inf:
    ge700K = eval(inf.read())
    
with open('data/800K/Ge/sites.dat', 'r') as inf:
    ge800K = eval(inf.read())
    
with open('data/900K/Ge/sites.dat', 'r') as inf:
    ge900K = eval(inf.read())
    
with open('data/1000K/Ge/sites.dat', 'r') as inf:
    ge1000K = eval(inf.read())
```

```python
x = [500, 600, 700, 800, 900, 1000]
w = [500, 600, 700, 800, 900, 1000]

y1 = [sn500K['Na1'], sn600K['Na1'], sn700K['Na1'], sn800K['Na1'], sn900K['Na1'], sn1000K['Na1']]
z1 = [ge500K['Na1'], ge600K['Na1'], ge700K['Na1'], ge800K['Na1'], ge900K['Na1'], ge1000K['Na1']]

y2 = [sn500K['Na2'], sn600K['Na2'], sn700K['Na2'], sn800K['Na2'], sn900K['Na2'], sn1000K['Na2']]
z2 = [ge500K['Na2'], ge600K['Na2'], ge700K['Na2'], ge800K['Na2'], ge900K['Na2'], ge1000K['Na2']]

y3 = [sn500K['Na3'], sn600K['Na3'], sn700K['Na3'], sn800K['Na3'], sn900K['Na3'], sn1000K['Na3']]
z3 = [ge500K['Na3'], ge600K['Na3'], ge700K['Na3'], ge800K['Na3'], ge900K['Na3'], ge1000K['Na3']]

y4 = [sn500K['Na4'], sn600K['Na4'], sn700K['Na4'], sn800K['Na4'], sn900K['Na4'], sn1000K['Na4']]
z4 = [ge500K['Na4'], ge600K['Na4'], ge700K['Na4'], ge800K['Na4'], ge900K['Na4'], ge1000K['Na4']]

y5 = [sn500K['Na5'], sn600K['Na5'], sn700K['Na5'], sn800K['Na5'], sn900K['Na5'], sn1000K['Na5']]
z5 = [ge500K['Na5'], ge600K['Na5'], ge700K['Na5'], ge800K['Na5'], ge900K['Na5'], ge1000K['Na5']]

y6 = [sn500K['Na6'], sn600K['Na6'], sn700K['Na6'], sn800K['Na6'], sn900K['Na6'], sn1000K['Na6']]
z6 = [ge500K['Na6'], ge600K['Na6'], ge700K['Na6'], ge800K['Na6'], ge900K['Na6'], ge1000K['Na6']]

y7 = [sn500K['i2'], sn600K['i2'], sn700K['i2'], sn800K['i2'], sn900K['i2'], sn1000K['i2']]
z7 = [ge500K['i2'], ge600K['i2'], ge700K['i2'], ge800K['i2'], ge900K['i2'], ge1000K['i2']]

# plt.subplot(711)
f, axs = plt.subplots(7,1,figsize=(8,10),sharex=True, gridspec_kw = {'wspace':0, 'hspace':0})
axs[0].plot(x, y1, label='Sn')
axs[0].plot(w, z1, label='Ge')
axs[0].set_title('Sulfide Average Site Occupancy with Temperature')
axs[0].set_ylabel('Na1')
axs[0].legend()
# plt.ylim(0,1)

# plt.subplot(712)
axs[1].plot(x, y2, w, z2)
axs[1].set_ylabel('Na2')
# plt.ylim(0,1)


axs[2].plot(x, y3, w, z3)
axs[2].set_ylabel('Na3')
# plt.ylim(0,1)


axs[3].plot(x, y4, w, z4)
axs[3].set_ylabel('Na4')
# plt.ylim(0,1)


axs[4].plot(x, y5, w, z5)
axs[4].set_ylabel('Na5')
# plt.ylim(0,1)


axs[5].plot(x, y6, w, z6)
axs[5].set_ylabel('Na6')
# plt.ylim(0,1)


axs[6].plot(x, y7, w, z7)
axs[6].set_xlabel('Temperature / K')
axs[6].set_ylabel('i2')
# plt.ylim(0,1)

# plt.savefig('S_Na_occupancies.png', dpi=300)
plt.show()
```
